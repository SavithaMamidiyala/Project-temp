```python
#Calculate number of inpatients
import datetime
from datetime import timedelta
inpatient_variable = variable6.loc[:,['SUBJECT_ID','HADM_ID','ADMITTIME','ADMISSION_TYPE']]
#inpatient_variable
inpatient_variable = inpatient_variable[inpatient_variable.ADMISSION_TYPE.str.lower() != 'emergency']
#inpatient_variable.count()
inpatient_var = inpatient_variable.groupby(['SUBJECT_ID'])['ADMITTIME'].max().reset_index(name='max_admit_time')
inpatient_var['max_admit_time'] = pd.to_datetime(inpatient_var.max_admit_time)
inpatient_var['Admit_Range'] = inpatient_var.max_admit_time - pd.offsets.DateOffset(years=1)
#inpatient_var.head()
nip_mrg = inpatient_variable.merge(inpatient_var, on = 'SUBJECT_ID', how='left')
nip_mrg['ADMITTIME'] = pd.to_datetime(nip_mrg.ADMITTIME)
#nip_mrg
inpatient_var_final = nip_mrg[(nip_mrg['ADMITTIME'] >= nip_mrg['Admit_Range']) & (nip_mrg['ADMITTIME'] 
                                                                                  <= nip_mrg['max_admit_time'])]
inpatient_var_final = inpatient_var_final.sort_values(['SUBJECT_ID','HADM_ID'],ascending = True)
#inpatient_var_final.count()
num_inpat_variable = inpatient_var_final.groupby(['SUBJECT_ID'])['HADM_ID'].count().reset_index(name='num_inpatient')
num_inpat_variable.head()
```

```python
#Calculate number of emergency patients
import datetime
from datetime import timedelta
emerpatient_variable = variable6.loc[:,['SUBJECT_ID','HADM_ID','ADMITTIME','ADMISSION_TYPE']]
emerpatient_variable = emerpatient_variable[emerpatient_variable.ADMISSION_TYPE.str.lower() == 'emergency']
emerpatient_var = emerpatient_variable.groupby(['SUBJECT_ID'])['ADMITTIME'].max().reset_index(name='max_admit_time')
emerpatient_var['max_admit_time'] = pd.to_datetime(emerpatient_var.max_admit_time)
emerpatient_var['Admit_Range'] = emerpatient_var.max_admit_time - pd.offsets.DateOffset(years=1)
#emerpatient_var.head()
nep_mrg = emerpatient_variable.merge(emerpatient_var, on = 'SUBJECT_ID', how='left')
nep_mrg['ADMITTIME'] = pd.to_datetime(nep_mrg.ADMITTIME)
#nep_mrg
emerpatient_var_final = nep_mrg[(nep_mrg['ADMITTIME'] >= nep_mrg['Admit_Range']) & 
                                (nep_mrg['ADMITTIME'] <= nep_mrg['max_admit_time'])]
#emerpatient_var_final
emerpatient_var_final = emerpatient_var_final.sort_values(['SUBJECT_ID','HADM_ID'],ascending = True)
#emerpatient_var_final
num_emer_variable = emerpatient_var_final.groupby(['SUBJECT_ID'])['HADM_ID'].count().reset_index(name='num_emergency')
num_emer_variable.head()
```

```python
#Calculate readmitted 
import datetime
from datetime import timedelta
readmit_variable = variable6.loc[:,['SUBJECT_ID','HADM_ID','ADMITTIME','DISCHTIME']]
#readmit_var = readmit_variable.groupby(['SUBJECT_ID','DISCHTIME'])
readmit_variable['ADMITTIME'] = pd.to_datetime(readmit_variable.ADMITTIME)
readmit_variable['DISCHTIME'] = pd.to_datetime(readmit_variable.DISCHTIME)
readmit_variable['Admit_Range'] = readmit_variable.DISCHTIME + pd.offsets.DateOffset(months=1)
readmit_variable = readmit_variable.sort_values(['SUBJECT_ID','HADM_ID'],ascending = True)
 
readmit_variable['ADMITTIME_s']= readmit_variable['ADMITTIME'].shift(-1)
readmit_variable['SUBJECT_ID_s']= readmit_variable['SUBJECT_ID'].shift(-1).fillna(0).astype(int)
readmit_var_final = readmit_variable[(readmit_variable['ADMITTIME_s'] >= readmit_variable['DISCHTIME']) & 
                                    (readmit_variable['ADMITTIME_s'] <= readmit_variable['Admit_Range']) & 
                                     (readmit_variable['SUBJECT_ID_s'] == readmit_variable['SUBJECT_ID'])]
readmit_var_final['Readmit'] = readmit_var_final['SUBJECT_ID'] == readmit_var_final['SUBJECT_ID_s']

readmit_var = readmit_var_final.loc[:,['SUBJECT_ID','HADM_ID','Readmit']]
#readmit_variable
readmit_var.head()
```
